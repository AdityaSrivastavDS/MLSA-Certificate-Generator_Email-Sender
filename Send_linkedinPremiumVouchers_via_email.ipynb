```python
# function to get winners from the list of participants at random
import random
import time

quota = 10

with open('./Data/Participant List.csv', 'r') as f:
    participant_list = f.readlines()
    participant_list = [x.strip().split(',') for x in participant_list]
    participant_list = participant_list[1:]

def get_random_string(participant_list):
    "Get current time as seed"
    current_time = int(time.time())
    "Use seed to initialize random number generator"
    random.seed(current_time)
    "Select random string from list"
    random_string = random.choice(participant_list)
    return random_string

for i in range(quota):
    print(get_random_string(participant_list))
    time.sleep(random.randint(1,10))
```
